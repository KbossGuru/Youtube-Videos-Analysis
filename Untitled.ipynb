```python
# importing the librabries 

import pandas as pd 
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the datasets...

canada= pd.read_csv('CAvideos.csv')
usa= pd.read_csv('USvideos.csv')
britain=pd.read_csv('GBvideos.csv')
germany= pd.read_csv('DEvideos.csv')
france= pd.read_csv('FRvideos.csv')
india= pd.read_csv('INvideos.csv')

#aadding a column for country in each dataset..
canada['Country']='CA'
usa['Country']='US'
britain['Country']='GB'
germany['Country']='DE'
france['Country']='FR'
india['Country']='IN'

#combining all the datasets
first_merge= pd.merge(canada, usa, how='outer')
second_merge= pd.merge(first_merge, britain, how='outer')
third_merge= pd.merge(second_merge, germany, how='outer')
fourth_merge= pd.merge(third_merge, india, how='outer')
df= pd.merge(fourth_merge, france, how='outer')
```

```python
# fixing the datatypes of each columns


#video_id
df['video_id']= df['video_id'].astype('str')
    
#title
df['title']= df['title'].astype('str')
    
#channel_title
df['channel_title']= df['channel_title'].astype('str')
    
#category_id
df['category_id']= df['category_id'].astype('str')
    
#publish_time
df['publish_time']= pd.to_datetime(df['publish_time'], errors= 'coerce', format='%Y-%m-%dT%H:%M:%S.%fZ')
    
#tags
df['tags']= df['tags'].astype('str')
    
#thuumbnail_link
df['thumbnail_link']=df['thumbnail_link'].astype('str')
    
#change comments_disabled, ratings_disabled, video_error_or_removed from bool to categorical datatype
df['comments_disabled']= df['comments_disabled'].astype('category')
df['ratings_disabled']= df['ratings_disabled'].astype('category')
df['video_error_or_removed']= df['video_error_or_removed'].astype('category')    


df.head(100)
```

```python
#seperate the trending_date into year month and day 
year=[]
for item in df['trending_date']:
    year.append(str(int(item.split('.')[0])+ 2000))

df['year']= pd.Series(year)

day=[]
for item in df['trending_date']:
    day.append(item.split('.')[1])
df['day']= pd.Series(day)

month=[]
for item in df['trending_date']:
    month.append(item.split('.')[2])
df['month']= pd.Series(month)

del df['trending_date']

df['trending_date']= pd.to_datetime(df[['year', 'month', 'day']].agg('-'.join, axis=1), format='%Y-%m-%d')
del df['year']
del df['month']
del df['day']
    
    
df.head()
```

```python
# seperate the publish_time into date and time)
df['publish_date']= df['publish_time'].dt.date
df['publish_time']= df['publish_time'].dt.time

df.head()
```

```python
#make a backup dataframe

backup_df= df.reset_index().sort_values('trending_date', ascending= False).set_index('video_id')

#sorting accoording to latest trending video and removing duplicates

df= df.reset_index().sort_values('trending_date', ascending= False).drop_duplicates('video_id', keep='first').set_index('video_id')

df.head()
```

```python
#reading the json file

with open('US_category_id.json', 'r') as f:
    data= f.read()
    
#parse data

obj=  json.loads(data)
```

```python
# linking the data of category_id into our dataframe

category_id= {}

with open('DE_category_id.json','r')as f:
    d= json.load(f)
    for category in d['items']:
        category_id[category['id']]= category['snippet']['title']
        
df.insert(2, 'category', df['category_id'].map(category_id))
backup_df.insert(2, 'category', df['category_id'].map(category_id))

df.head()
```

```python
#finding the ratio of likes to dislikes in different categories

likesdf= df.groupby('category')['likes'].agg('sum')

dislikesdf= df.groupby('category')['dislikes'].agg('sum')

ratiodf= likesdf/dislikesdf

ratiodf= ratiodf.sort_values(ascending= False).reset_index()


plt.figure(figsize=(10,12))
sns.barplot(x=0, y='category', data=ratiodf, label= 'Likes-To-Dislikes Ratio', color='b')
plt.show()
```

```python
#showing the category the users likes the most

likesdf= df.groupby('category')['likes'].agg('sum')
likesdf= likesdf.sort_values(ascending=False).reset_index()

plt.figure(figsize=(10,10))
sns.barplot(x='likes', y='category',data=likesdf, label='Likes',  color='g')
plt.show()

#output shows that among all the countries shows that movies, shows and trailers were the least liked.
```
